【問題1】乱数の作成

```python
import numpy as np
mu=[-3,0]
sigma=[[1.0, 0.8],
       [0.8, 1.0]]
np.random.seed(0)
ndarray=np.random.multivariate_normal(mu,sigma,500)
print(ndarray)
```

【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(ndarray[:,0],ndarray[:,1])
plt.show()
```

【問題3】ヒストグラムによる可視化

```python
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")
plt.hist(ndarray[:,0])
plt.xlim(-6,3)
plt.show()
```

```python
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.hist(ndarray[:,1],color='green')
plt.xlim(-6,3)
plt.show()
```

【問題4】データの追加　※問題１と区別するべく、乱数シードを１とする。

```python
mu1 = [0,-3]
sigma1 = [[1.0, 0.8],
       [0.8, 1.0]]
np.random.seed(1)
ndarray1 = np.random.multivariate_normal(mu1,sigma1,500)
#print(ndarray1) #表示は省略する。

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(ndarray[:,0],ndarray[:,1],label="0")
plt.scatter(ndarray1[:,0],ndarray1[:,1],color='orange',label="1")
plt.legend()
plt.show()
```

【問題5】データの結合

```python
ndarray_merge = np.concatenate((ndarray,ndarray1),axis=0) #axis=0は列方向。デフォルトでaxis=0
print(ndarray_merge)
```
