##  問題２

```python
def get_derivative(mount_data):
#Ⅹ方向の差分を求める
    position = mount_data[:,0]
    position_0 = position[1:]
    position_1 = position[:-1]
    position_dif = position_0 - position_1

#Ｙ方向の差分を求める
    hight = mount_data[:,3]
    hight_0 = hight[1:]
    hight_1 = hight[:-1]
    hight_dif = hight_0 - hight_1
    return hight_dif/position_dif

get_derivative(fuji)
```

## 問題３

```python
import numpy as np
def move_destination(i_position):
    derivative_ndarray = get_derivative(fuji)
    position_derivative = derivative_ndarray[i_position]
    destination = i_position - 0.2 * position_derivative
    # 移動後の位置が負になるのを防ぐための条件分岐。
    if destination < 0:
       destination = i_position - 0.05 * position_derivative
    return np.round(destination)

# derivative_ndarray[i_position]は、地点i_positionにおける傾きを表しているのか？

print(move_destination(136))
for i in range(299):
   print(move_destination(i))
```

## 問題４

```python
# trace_data（リスト）に、移動後の位置を繰り返し処理（gezan関数）で順次格納する。
# gezan関数では、繰り返し処理を1000回実行して挙動をみる（位置７４で下山終了となる）。

import numpy as np

def gezan(i_position):
    trace_data = []
    trace_data.append(i_position)
    for i in range(1000):
        temp = move_destination(trace_data[i])
        temp = int(temp)
        trace_data.append(temp)
    return trace_data

gezan(136)
```

## 問題５

```python
x_data = gezan(136)
y_data = []

for i in x_data:
    y_data.append(fuji[i,3])

plt.title("Mt.Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(fuji[:, 0], fuji[:, 3])
plt.scatter(x_data,y_data,color='r')
plt.show()
```

# 問題7

```python
# hyperパラメータを引数として設定できるようにするため、
# gezan関数とmove_desitination関数を再定義（gezan7とmove_desitination7）

def move_destination7(i_position,hyper):
    derivative_ndarray = get_derivative(fuji)
    position_derivative = derivative_ndarray[i_position]
    destination = i_position - hyper * position_derivative
    # 移動後の位置が負になるのを防ぐための条件分岐。
    if destination < 0:
       destination = i_position - 0.05 * position_derivative
    return np.round(destination)

def gezan7(i_position, hyper = 0.2):
    trace_data = []
    trace_data.append(i_position)
    for i in range(1000):
        temp = move_destination7(trace_data[i], hyper)
        temp = int(temp)
        trace_data.append(temp)
    return trace_data

# 初期位置136、ハイパーパラメータ0.02
x_data = gezan7(136,0.02)
y_data = []

for i in x_data:
    y_data.append(fuji[i,3])

plt.title("Mt.Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(fuji[:, 0], fuji[:, 3])
plt.scatter(x_data,y_data,color='r')
plt.show()
```
