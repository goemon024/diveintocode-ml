# 問題５　歪度と尖度について

　歪度は、平均値μまわりの３次モーメントをＥ（（ｘ－μ）<sup>３</sup>）を、標準偏差の３乗（σ<sup>３</sup>）で除して標準化したもの。  
　尖度は、平均値μまわりの４次モーメントをＥ（（ｘ－μ）<sup>３</sup>）を、標準偏差の４乗（σ<sup>３</sup>）で除して標準化して３を引いたもの。
　歪度は、正規分布を基準としての左右対称性の偏りを示す指標化したものであり、尖度は、正規分布を基準としての裾の広がり度合いとピークの突出度合いを指標化したものである。  
　歪度は、３次モーメントであるために平均値回りの左右ずれ量が正負で表出されるものであり（左右の偏りが大きいほど、歪度がゼロから離れる）、尖度は、４次モーメントであるために平均値から遠い部分が寄与しやすくなる（すそ野が広がるほど、尖度は大きくなる）。

# 問題６　分布の確認

```python
import seaborn as sns
sns.distplot(aims_data["SalePrice"])
print("歪度:{}".format(aims_data['SalePrice'].skew()))
print("尖度:{}".format(aims_data['SalePrice'].kurt()))
```

```python
#対数変換
import numpy as np
saleprice_log = np.log(aims_data["SalePrice"])
sns.distplot(saleprice_log)

print("対数変換後の歪度:{}".format(saleprice_log.skew()))
print("対数変換後の尖度:{}".format(saleprice_log.kurt()))
```

＜説明＞　対数変換をすることで、歪度も尖度もゼロに近づいた。  
すなわち対数変換により、左に偏って右側に裾が広がっていた元々の分布が、正規分布に近づいた（平均値付近にピークを有して、分布の広がり度合いとピークの突出度合いが抑制された）。

```python
aims_data_corr = aims_data.corr()
temp = aims_data_corr["SalePrice"]

#相関係数0.5以上の説明変数をtemp1リストに格納する。
temp1 = []
for key,value in temp.items():
    if value > 0.50:
       temp1.append(key)

#格納された10個の説明変数のヒートマップを作成。
temp2 = aims_data_corr[temp1]

plt.figure(figsize=(8,8))
sns.set_palette("husl")
sns.heatmap(temp2.corr(),vmax=1.0, vmin=-1.0, square=True, cbar_kws={"shrink": .5})
```
