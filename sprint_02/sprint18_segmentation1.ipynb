## 【問題２】　コードリーディング

### 論文について
[1]O.Ronneberger, P.Fischer, T.Brox. U-Net: Convolutional Networks for Biomedical Image Segmentation.   
Medical Image Computing and Computer-Assisted Intervention (MICCAI), Springer, LNCS, Vol.9351: 234–241, 2015

### 論文の総括（結論の和訳）
Unetを用いることで医療分野でのセグメンテーションで良い結果を達成できた。
オーグメンテーションのおかげで、非常にわずかなアノテーションされたデータによる計算で良好な結果が得られる（6GBのGPUで計算時間10時間ぐらい）。Unetは色々なタスクに応用できるでしょう。

### 論文の概要（イントロの和訳がメイン）
・すべてＣＮＮによるネットワーク構成で、入力画像を訓練してセグメンテーションされた出力画像を生み出す。  
・ネットワークの前半は、エンコーダーであって畳み込み層＋max pooling層で構成されて、入力画像からの特徴量を抽出を行う。  
・ネットワークの後半は、デコーダーであって、畳み込み層＋upsampling層で構成されて、特徴マップに位置特定情報を組み込んでいく。  
  
・パッチサイズの大小は、位置特定性能と特徴抽出性能のトレードオフとなる。バッチサイズを大きくすると、多くのマックスプーリング層が必要になり位置特定の正確さが失われる。パッチサイズを小さくすると、わずかな特徴しかとらえられなくなる。  
・このようなトレードオフに対して、特徴マップを複数層から取り込むことで改善する（encorderとdecorderの間にある「contracting network」≒Resnetにおけるskip connection。参考：https://qiita.com/koshian2/items/603106c228ac6b7d8356）。  

###  github上のコードと、論文のイントロの記載との対比。  
https://github.com/zhixuhao/unet

merge6 = concatenate([drop4,up6], axis = 3)
