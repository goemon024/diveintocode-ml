# 【問題8】可視化したグラフや表の結果の説明

petal:花びら。sepal:がく。 
ラベル０：setosa,　ラベル１：virgiclor,　ラベル２：virginca 

１．petal_lengthとpetal_widthの相関係数は0.96であり、花びらの長さと幅は比例の関係にある。  
 すなわち花びらの長さが大きくなれば、幅もそれに応じて大きくなる関係にある。  
 また、ラベル１、ラベル２と、ラベル３の花びらの大きさは、ラベル２＞ラベル１＞ラベル０の関係にある。
 
２．ラベル０は、花びらの長さや幅が他のラベルよりも、明らかに小さい傾向にある。  
 ラベル０のガクは、その長さが他のラベルよりも小さい傾向にあるが、その幅が他のラベルよりも同等以上になっていてばらつきも大きい。  
 したがってラベル０は、花全体的には他よりも小さくなるがガクの幅の大きさがばらつく傾向にある。また、花全体のバランスとしてガクの幅が他よりも広がっているという特徴を持ちやすい。

３．ラベル１と２では、花びらの長さと幅で区別しやすい。  
  おおむね、花びらの長さが５ｃｍ以上ではラベル２となり、５ｃｍ未満ではラベル１となる。
  また概ね、花びらの幅が１．７５ｃｍ以上ではラベル２となり、１．７５ｃｍ未満ではラベル１となるが、ラベル２はばらつきも大きいようで、１．７５ｃｍ未満でもラベル２となる場合がある。
　ガクの長さや幅は、ラベル２がラベル１よりも大きくなる傾向にありそうだが同等のサイズであるサンプルも多い。したがってラベル１とラベル２を比較すると、ガクの大きさは同等であるものの花びらの大きさがラベル２の方が大きくなるということになる。
