【問題１】100日目の米粒の数  

```python
#変数宣言
rice = 0
sum_rice = 0

#空リストの作成
rice_list = [0]
sum_rice_list = [0]

for i in range(100):
    rice = 2**i
    sum_rice += rice
    rice_list.append(rice)
    sum_rice_list.append(sum_rice)
    
print("100日目に貰う米の数:{}粒".format(rice_list[100]))
print("100日目までの米粒の累計:{}粒".format(sum_rice_list[100]))

#チェック用
#print(rice_list)

import matplotlib.pyplot as plt
%matplotlib inline
plt.title("increasement of rice grains")
plt.xlabel("number of days")
plt.ylabel("number of grains of rice ")
plt.tick_params(labelsize=20, color='red')
plt.plot(rice_list,color=(1,0,1),label="day") 
plt.plot(sum_rice_list,color=(0,1,1),label="sumption") 
plt.legend()
plt.show()
```

【問題２】１００日目以外の米粒の数

```python
#一日に貰う米粒量、トータル米粒量を記録するためのリストを定義。
list_n_grains = []
list_total_grains = []

'''
米粒をもらう日にちを引き数、
「1日の米粒量リスト」と「貰った米粒の合計量リスト」を返り値
とする関数を定義する。riceとsum、list_a,list_bはinner変数、リスト。
'''
def compute_sorori_shinzaemon(day=10):
    rice = 0
    sum = 0
    list_a = [0]
    list_b = [0]
    for i in range(day):
        rice = 2**i
        sum += rice
        list_a.append(rice)
        list_b.append(sum)
    return list_a,list_b

#関数で返された結果をリストに入れる。求めたい日数を関数の引数として直接入力する。
day=30
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day)
print("貰える米粒量：{}粒".format(list_total_grains[day]))

import matplotlib.pyplot as plt
%matplotlib inline
plt.title("increasement of rice grains")
plt.xlabel("number of days")
plt.ylabel("number of grains of rice ")
plt.tick_params(labelsize=20, color='red')
plt.plot(list_n_grains,color=(1,0,1),label="amount of day") 
plt.plot(list_total_grains,color=(0,1,1),label="total") 
plt.legend()
plt.show()
```

【問題３】何人が毎日生活できる米か

成人1人が1日に必要とする米は4.5合。1合は概ね6483粒（6500粒）とされている。

```python
day = 29
person_number = 7
day_to_live = 0

list_n_grains, list_total_grains = compute_sorori_shinzaemon(day)

# compute_sorori_shinzaemon(day)に任意の日数を入力し、米粒合計量を4.5×6483で割る。

day_to_live = list_total_grains[day]/(person_number*4.5*6483)
print("貰える米粒量：{}粒".format(list_total_grains[day]))
print("{}人が、生活可能な日数：{:.0f}日".format(person_number,day_to_live))
```
